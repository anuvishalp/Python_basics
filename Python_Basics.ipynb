```python
""" Complete overview of dictionaries in python

Basic dictionary

person = {"name": "Alice", "age": 30, "city": "New York"}

✔️ Using dict()

person = dict(name="Alice", age=30, city="New York")

✔️ Empty dictionary
d = {}
d = dict()

🔑 2. Accessing Values

✔️ Using key
print(person["name"])

✔️ Safe access using .get()

print(person.get("salary"))        # returns None
print(person.get("salary", 0))     # default value

✏️ 3. Updating Values
person["age"] = 31

➕ 4. Adding New Key–Value Pairs

person["job"] = "Engineer"

❌ 5. Deleting Keys
✔️ del
del person["city"]

✔️ .pop()
age = person.pop("age")   # returns removed value

✔️ .popitem() — removes last inserted pair
person.popitem()

🔄 6. Dictionary Methods (Complete List with Examples) ->Here is the full set of commonly used dictionary methods.

📌 1. .keys() — returns all keys
print(person.keys())

📌 2. .values() — returns all values
print(person.values())

📌 3. .items() — returns key–value pairs
print(person.items())

📌 4. .update() — merge dictionaries
person.update({"country": "USA", "age": 32})

📌 5. .get() — safe key access
person.get("salary", "Not Available")

📌 6. .pop() — remove key and return value
removed = person.pop("job")

📌 7. .popitem() — remove last inserted pair
person.popitem()

📌 8. .setdefault() — get value OR set default
person.setdefault("hobby", "Reading")

📌 9. .clear() — remove all items
person.clear()

🧠 7. Looping Through Dictionaries

✔️ Loop keys
for key in person:
    print(key)

✔️ Loop values
for value in person.values():
    print(value)

✔️ Loop key–value pairs
for key, value in person.items():
    print(key, value)

🧩 8. Dictionary Comprehensions
✔️ Basic
squares = {x: x*x for x in range(5)}

✔️ Filtered
even_squares = {x: x*x for x in range(10) if x % 2 == 0}

🏗️ 9. Nested Dictionaries
✔️ Example

students = {
    "A101": {"name": "Alice", "grade": "A"},
    "A102": {"name": "Bob", "grade": "B"}
}
Access nested values

students["A101"]["grade"]

🔄 10. Merging Dictionaries
✔️ Python 3.9+
d3 = d1 | d2

✔️ Using .update()
d1.update(d2)

🧪 11. Common Real‑World Scenarios
🌐 Scenario 1: JSON from API
response = {
    "status": "success",
    "data": {"id": 101, "name": "Alice"}
}

print(response["data"]["name"])

🧾 Scenario 2: Counting frequency
freq = {}
for char in "banana":
    freq[char] = freq.get(char, 0) + 1

📊 Scenario 3: Grouping values
groups = {}
for city, country in [("Paris", "France"), ("Lyon", "France"), ("Berlin", "Germany")]:
    groups.setdefault(country, []).append(city)

🚗 Scenario 4: Car Dictionary (from your image)
car = {
    "make": "Toyota",
    "model": "Camry",
    "year": 2020,
    "color": "Red"
}

print(car["year"])
car["color"] = "Blue"
car["price"] = 25000

🧩 12. Dictionary vs List vs Set — Quick Comparison
Feature	Dictionary	List	Set
Ordered	Yes (3.7+)	Yes	No
Mutable	Yes	Yes	Yes
Duplicates	Keys: No, Values: Yes	Yes	No
Access	By key	By index	Not indexable
Use case	Structured data	Sequences	Unique items """
```
